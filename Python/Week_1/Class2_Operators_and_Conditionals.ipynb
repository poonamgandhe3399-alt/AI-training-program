## 2. Comparison Operators

Comparison operators **compare two values** and always return `True` or `False`.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `7 > 3` | `True` |
| `<` | Less than | `2 < 1` | `False` |
| `>=` | Greater than or equal | `5 >= 5` | `True` |
| `<=` | Less than or equal | `4 <= 3` | `False` |

⚠️ **Common mistake:** `=` is assignment, `==` is comparison!

````python
age = 20

print(age == 20)   # True  — is age exactly 20?
print(age != 18)   # True  — is age not 18?
print(age > 18)    # True  — is age greater than 18?
print(age < 18)    # False — is age less than 18?
print(age >= 20)   # True  — is age at least 20?
print(age <= 20)   # True  — is age at most 20?
````

## 3. Logical Operators

Combine multiple comparisons with , `not`, `and`, `or`:

|Priority Order | Operator | Meaning | Returns True when |
|-----|----------|---------|-------------------|
|1 | `not` | Flip True↔False | The condition is False |
|2 | `and` | Both must be True | Left AND right are True |
|3 | `or` | At least one True | Left OR right (or both) are True |


````python
print(True and False or True)# -> False or True -> True
print(not True or False)# -> False
````

````python
# AND Gate : Input ->  AND Gate -> Output
# Input_1 and Input_2 -> Output
# 1 and 1 -> 1, 1 and 0 -> 0, 0 and 1 -> 0, 0 and 0 -> 0


# Input_1 or Input_2 -> Output
# 1 or 1 -> 1, 1 or 0 -> 1, 0 or 1 -> 1, 0 or 0 -> 0

print(not False)
````

````python
age = 22
has_id = True
is_vip = False

# and — both must be True
can_enter = (age >= 18) and has_id
print("Can enter:", can_enter)   # True

# or — at least one True
gets_discount = age < 18 or is_vip
print("Gets discount:", gets_discount)   # False

# not — flip the result
is_minor = not (age >= 18)
print("Is minor:", is_minor)   # False
````

````python
# Truth table — understanding and/or
print("=== AND ===")
print(True  and True)    # True
print(True  and False)   # False
print(False and True)    # False
print(False and False)   # False

print("\n=== OR ===")
print(True  or True)     # True
print(True  or False)    # True
print(False or True)     # True
print(False or False)    # False

print("\n=== NOT ===")
print(not True)          # False
print(not False)         # True
````

## 4. Membership & Identity Operators

| Operator | Use | Example |
|----------|-----|---------|
| `in` | Is value inside a collection? | `"a" in "cat"` → `True` |
| `not in` | Is value NOT inside? | `5 not in [1,2,3]` → `True` |
| `is` | Same object in memory? | `x is None` → `True/False` |
| `is not` | Different objects? | `x is not None` → `True/False` |

````python
# in / not in
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)       
print("mango" in fruits)       
print("mango" not in fruits)  

# Works on strings too
email = "alice@gmail.com"
print("@" in email)            # True
print("gmail" in email)        # True

# is / is not — mostly used for None checks
result = None
print(result is None)          # True — preferred over result == None
print(result is not None)      # False
````

## 5. Conditional Statements — if / elif / else

This is how Python makes **decisions**:

```python
if condition:
    # runs when condition is True
elif another_condition:
    # runs when first is False but this is True
else:
    # runs when ALL conditions above are False
```

⚠️ **Indentation matters!** Python uses 4 spaces (or 1 tab) to define blocks.

````python
student_height = 4.10

# categories = 'short', 'medium', 'tall'


short = 4
tall = 5

if student_height < short: # 4.10 < 4 = False
    print('short')
elif short < student_height < tall: # 4 < 4.10 < 5 True
    print('medium')
else:
    print('tall')


````

````python
# Basic if/else
temperature = 35

if temperature > 30:
    print("It's hot today! Stay hydrated. 🌡️")
else:
    print("The weather is pleasant.")
````

````python
# if / elif / else — multiple branches
score = 82

if score >= 90:
    grade = "A"
    comment = "Excellent!"
elif score >= 80:
    grade = "B"
    comment = "Very good!"
elif score >= 70:
    grade = "C"
    comment = "Good"
elif score >= 60:
    grade = "D"
    comment = "Needs improvement"
else:
    grade = "F"
    comment = "Please retake"

print(f"Score: {score} → Grade: {grade} ({comment})")
````

````python
# Real-world example: loan eligibility
age = 28
income = 50000
credit_score = 720
has_existing_loan = False

if age < 18:
    print("❌ Must be 18 or older to apply")
elif income < 30000:
    print("❌ Minimum income requirement not met")
elif credit_score < 650:
    print("❌ Credit score too low")
elif has_existing_loan:
    print("⚠️ Existing loan detected — conditional approval")
else:
    print("✅ Congratulations! You are eligible for a loan.")
````

## 6. Nested Conditionals

`if` statements can go **inside** other `if` statements:

````python
# Nested if — be careful not to go too deep!
is_logged_in = True
is_admin = False
has_permission = True

if is_logged_in:
    print("✅ User is logged in")
    if is_admin:
        print("👑 Admin panel unlocked")
    else:
        if has_permission:
            print("📄 Access to permitted sections")
        else:
            print("🔒 Limited access only")
else:
    print("❌ Please log in first")
````

## 7. Ternary (One-line) Conditionals

For simple True/False assignments, Python has a clean one-liner:

```python
value_if_true  if  condition  else  value_if_false
```

````python
# Standard way
age = 20
if age >= 18:
    status = "adult"
else:
    status = "minor"
print(status)

# Ternary way — same result, one line
status = "adult" if age >= 18 else "minor"
print(status)

# More examples
score = 85
result = "Pass" if score >= 50 else "Fail"
print(result)

number = 7
parity = "even" if number % 2 == 0 else "odd"
print(f"{number} is {parity}")
````
