## 1. What is a Variable?

A **variable** is a named container that stores a value in your computer's memory.

Think of it like a labelled box 📦:
- The **label** is the variable name
- The **contents** are the value
- You can swap the contents anytime

```
name = "Alice"
  ↑       ↑
label   contents
```

### Rules for naming variables
| Rule | ✅ Good | ❌ Bad |
|------|---------|--------|
| Letters, digits, underscores only | `user_age` | `user-age` |
| Cannot start with a digit | `score1` | `1score` |
| Case-sensitive | `Name ≠ name` | — |
| No Python keywords | `total` | `for`, `if` |
| Use snake_case by convention | `first_name` | `firstName` |

````python
# Creating your first variables
name = "Alice"          # storing text
age  = 25               # storing a whole number
height = 5.6            # storing a decimal number
is_student = True       # storing True or False

# Print them all
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is student:", is_student)
````

## 2. The Five Core Data Types

Python has five fundamental data types every programmer must know:

| Type | Python name | Example | Used for |
|------|-------------|---------|----------|
| Integer | `int` | `42` | Whole numbers |
| Float | `float` | `3.14` | Decimal numbers |
| String | `str` | `"hello"` | Text |
| Boolean | `bool` | `True` | Yes/No, On/Off |
| None | `NoneType` | `None` | Absence of value |

````python
unknown = None
print(unknown, type(unknown))
````

````python
boolean = True
print(f"Boolean = {boolean}")

name = "Sai"
print("Name:", name, "boolean", boolean)
print(f"Name = {name} boolean = {boolean}") # f string method
````

````python
# --- int (integers) ---
students_count = 30
temperature_celsius = -5
year = 2024

print("Type of students_count:", type(students_count))   # <class 'int'>

# --- float (floating point numbers) ---
pi = 3.14159
gpa = 3.75
salary = 85000.50

print("Type of pi:", type(pi))   # <class 'float'>
````

````python
# --- str (strings = text) ---
first_name = "Alice"
last_name  = 'Smith'          # single OR double quotes — both work
greeting   = "Hello, World!"

print(first_name)
print(type(greeting))         # <class 'str'>

# Multi-line strings use triple quotes
bio = """Alice is a data scientist. 
         She loves Python and machine learning."""
print(bio)
````

````python
# --- bool (booleans) ---
is_raining   = False
has_passed   = True
account_active = True

print("Is raining?", is_raining)
print("Type:", type(is_raining))   # <class 'bool'>

# Booleans are actually True = 1 and False = 0 under the hood!
print(True + True)    # -> True -> 2
print(True + False)   # -> False -> 1
````

````python
# --- NoneType ---
# None means "no value" or "not yet assigned"
result = None
user_input = None

print("result:", result)
print("Type:", type(result))   # <class 'NoneType'>

# Common use: a function that hasn't returned anything yet
def pending_function():
    pass   # does nothing yet
    return

output = pending_function()
print("Function returned:", output)   # None
````

## 4. Type Conversion (Casting)

You can **convert** between types using built-in functions:

| Function | Converts to | Example |
|----------|-------------|---------|
| `int()` | Integer | `int("42")` → `42` |
| `float()` | Float | `float(5)` → `5.0` |
| `str()` | String | `str(100)` → `"100"` |
| `bool()` | Boolean | `bool(0)` → `False` |

This is called **casting** or **type conversion**.

````python
# int() — convert to integer
x = int("42")      # string → int
y = int(3.99)      # float → int (truncates, does NOT round!)
z = int(False)      # bool  → int

print(x, type(x))
print(y, type(y))  # 3, not 4 !
print(z, type(z))  # 1
````

````python
x = 10%2.5 # 10 = quotient*divisor + reminder
print(x)
````

````python
# float() — convert to float
a = float("3.14")   # string → float
b = float(10)       # int    → float

print(a, type(a))
print(b, type(b))   # 10.0
````

````python
# str() — convert to string (very common for printing)
score = 95
message = "Your score is: " + str(score)   # must convert before concatenating
print(message)

# bool() rules:
# Falsy: 0, 0.0, "", None, [], {}
# Truthy: everything else
print(bool(0))      # False
print(bool(""))     # False
print(bool("hi"))   # True
print(bool(42))     # True
print(bool(None))   # False
````

````python
# ⚠️ Type errors — common mistakes
# This will CRASH — uncomment and try it:
# result = "Age: " + 25   # TypeError: can only concatenate str to str

# Correct way:
result = "Age: " + str(25)
print(result)

# Another crash — uncomment to see:
# x = int("hello")   # ValueError: can't convert "hello" to int
````

## 5. String Operations Preview

Strings have powerful built-in abilities — we'll go deep in Class 3, but here's a taste:

````python
name = "Alice"
city = "Hyderabad"

# Concatenation — joining strings
greeting = "Hello, " + name + "! Welcome to " + city + "."
print(greeting)

print("Hello, ", name , "! Welcome to " , city , ".")

# f-strings — the modern, clean way 

# String length
print("Length of name:", len(name))   # 5

# Upper / lower
print(name.upper())   # ALICE
print(name.lower())   # alice
````

## 6. Variable Reassignment & Multiple Assignment

Variables can change value — they are not fixed!

````python
# Reassignment
score = 0
print("Start:", score)

score = 50
print("After update:", score)

score = score + 10    # common pattern: update using current value
print("After +10:", score)

score -= 10           # shorthand for score = score - 10
print("After +=10:", score)
````

````python
# Multiple assignment — assign several at once
x, y, z = 1, 2, 3
print(x, y, z)

# Assign same value to multiple variables
a = b = c = 0
print(a, b, c)

# Swap values (Python magic — no temp variable needed!)
x, y = 10, 20
print(f"Before swap: x={x}, y={y}")
x, y = y, x
print(f"After swap:  x={x}, y={y}")
````

## 7. Constants Convention

Python has no built-in constant keyword, but by convention:
- Variables written in **ALL_CAPS** are treated as constants (don't change them!)

````python
# Constants — by convention, use ALL_CAPS
MAX_STUDENTS = 50
PI = 3.14159
APP_VERSION = "1.0.0"
DATABASE_URL = "localhost:5432"

print(f"Max students allowed: {MAX_STUDENTS}")
print(f"Pi ≈ {PI}")
````
