**Why functions?**
- Avoid repeating the same code (DRY: Don't Repeat Yourself)
- Break complex problems into small pieces
- Make code readable and testable

````python
temperatures = [20, 25, 32, 35, 45, 42, 47, 50, 52]

for temp in temperatures: # to iterate over an iterable
    print(f"recorded temp = {temp}")
    if temp <= 45:
        print("We are safe!")
    else:
        print(f"The temperature is beyond threshold {temp}, be carefull!")
        break
````

````python
temp = 0
while temp <= 45: #until condition holds
    print(f"We are safe! temp = {temp}")
    temp += 1 

print(f"The temperature is beyond threshold {temp}, be carefull!")
````
