## 1. Dictionaries

A **dictionary** stores **key-value pairs** — like a real dictionary where you look up a word (key) to get its definition (value).

```
{"name": "Alice",  "age": 25}
       ↑               ↑
     key:value      key:value
```

- Keys must be **unique** and **immutable** (strings, numbers, tuples)
- Values can be **anything** (strings, lists, other dicts, ...)
- Ordered since Python 3.7

````python
dict_list = [('name', 'Alice'), ('city', "Hyderabad"), ('age', 25)] #[(key, value), (key, value)]
                    #0                1             2

# traversing throught the list takes time
for key, value in dict_list:
    if key == 'age':
        print(f"Age = {value}")

# time complexity of searching through list = O(n)
````

### 1.1 Accessing & Modifying

````python
student = {"name": "Alice", "age": 22, "gpa": 3.8, "city": "Hyderabad"}

# Access by key
print(student["name"])    # Alice
print(student["gpa"])     # 3.8

# .get() — safe access (returns None if key missing, no crash)
print("Age",student.get("Age", 25))           # 22
print(student.get("phone"))         # None
print(student.get("phone", "N/A"))  # N/A — custom default

# Modify existing value
student["age"] = 23
print("Updated age:", student["age"])

# Add new key-value pair
student["email"] = "alice@example.com"
print("Email added:", student)

# Delete a key
del student["city"]
removed = student.pop("email")   # removes and returns value
print("Removed:", removed)
print("Final:", student)
````

### 1.2 Iterating over Dictionaries

````python
student = {"name": "Alice", "age": 22, "gpa": 3.8, "enrolled": True}

# Loop over keys (default)
print("=== Keys ===")
for key in student:
    print(key)

# Loop over values
print("\n=== Values ===")
for value in student.values():
    print(value)

# Loop over key-value pairs (most common)
print("\n=== Items ===")
for key, value in student.items():
    print(f"  {key}: {value}")
````

````python
# Useful dictionary methods
profile = {"name": "Bob", "age": 30, "city": "Mumbai"}

print("Keys:   ", list(profile.keys()))    # dict_keys → convert to list
print("Values: ", list(profile.values()))
print("Items:  ", list(profile.items()))

# Check if a key exists
print("name" in profile)      # True
print("salary" in profile)    # False

# Merge two dicts (Python 3.9+ uses |, earlier uses update)
extra = {"salary": 75000, "department": "Engineering"}
profile.update(extra)
print("Merged:", profile)
````

## 2. Sets

A **set** is an **unordered** collection of **unique** items.
Great for removing duplicates and fast membership testing.

```python
{1, 2, 3, 4}   # set
```

⚠️ Empty set = `set()` not `{}` — that makes an empty dict!

````python
fruits = set()

# Creating sets
fruits = {"apple", "banana", "cherry", "apple", "banana"}
print(fruits)           # duplicates removed automatically!
print(type(fruits))     # <class 'set'>

# From a list (common way to remove duplicates)
numbers_with_dupes = [5, 1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
unique_numbers = set(numbers_with_dupes)
print(unique_numbers)   # {1, 2, 3, 4}

# Convert back to sorted list
clean_list = sorted(list(unique_numbers))
print(clean_list)
````

````python
# Set operations — like Venn diagrams!
python_students = {"Alice", "Bob", "Charlie", "Diana"}
ml_students     = {"Bob", "Diana", "Eve", "Frank"}

# Union — all students in either course
print("Union:       ", python_students | ml_students)

# Intersection — students in BOTH
print("Intersection:", python_students & ml_students)

# Difference — in Python but NOT in ML
print("Difference:  ", python_students - ml_students)

# Symmetric difference — in one but NOT both
print("Sym. diff:   ", python_students ^ ml_students)

# Membership test — very fast even with millions of items
print("Alice in Python:", "Alice" in python_students) #O(1) not O(n)
````

## 3. Functions

A **function** is a named, reusable block of code.

**Why functions?**
- Avoid repeating the same code (DRY: Don't Repeat Yourself)
- Break complex problems into small pieces
- Make code readable and testable

```python
def function_name(parameters):
    """Docstring: explain what this function does"""
    # code
    return result
```

````python
# Your first function
def greet():
    """Print a simple greeting."""
    print("Hello, welcome to Python!")

# Call the function (nothing happens until you call it)
greet()
greet()   # call it again — that's the power of functions!
````

````python
# Function with parameters
def greet_person(name):
    """Greet a specific person by name."""
    print(f"Hello, {name}! Welcome to Python!")
    return None

greet_person("Alice")
greet_person("Bob")
greet_person("Charlie")
````

````python
# Function with return value
def add(a : int, b : int):
    """Add two numbers and return the result."""
    result = a + b
    return result

total = add(10, 25)
print("Sum:", total)
print("Direct:", add(3, 4))   # use return value directly
````

### 3.1 Default Parameters

````python
# Default parameter values
def greet(name, greeting="Hello"):
    """Greet with customisable greeting. Default is Hello."""
    print(f"{greeting}, {name}!")

greet("Alice")                  # uses default "Hello"
greet("Bob", "Good morning")    # overrides default
greet("Carol", greeting="Hi")   # using keyword argument
````

````python
# Keyword arguments — order doesn't matter when named
def create_profile(name, age, city="Unknown", role="Student"):
    return {
        "name": name,
        "age":  age,
        "city": city,
        "role": role
    }

# Different ways to call the same function:
p1 = create_profile("Alice", 22)
p2 = create_profile("Bob", 30, "Mumbai", "Engineer")
p3 = create_profile(age=25, name="Carol", role="Designer")
p4 = create_profile("Dan", 28, city="Hyderabad")

for p in [p1, p2, p3, p4]:
    print(p)
````

### 3.2 Multiple Return Values

````python
# Return multiple values (Python returns a tuple)
def stats(numbers):
    """Return min, max, and average of a list."""
    return min(numbers), max(numbers), sum(numbers) / len(numbers)

data = [10, 25, 4, 88, 42, 17, 63]

low, high, avg = stats(data)   # unpack the tuple
print(f"Min: {low}, Max: {high}, Average: {avg:.2f}")
````

### 3.3 *args and **kwargs — Flexible Functions

````python
# *args — accept any number of positional arguments
def total(a, *args):
    """Sum any number of values."""
    print(f"Received: {a}, {args}")   # args is a tuple
    return a, sum(args)

print(total(1, 2, 3))
print(total(10, 20, 30, 40, 50))
print(total(5))
````

````python
# **kwargs — accept any number of keyword arguments
def display_info(**kwargs):
    """Display any key-value info passed in."""
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

display_info(name="Alice", age=22, city="Hyderabad")
display_info(model="GPT-4", year=2023, company="OpenAI")
````

````python
# Combining all: positional, *args, keyword, **kwargs
def full_example(required, *args, keyword="default", **kwargs):
    print(f"Required: {required}")
    print(f"Extra positional (args): {args}")
    print(f"Keyword: {keyword}")
    print(f"Extra keyword (kwargs): {kwargs}")

full_example("hello", 1, 2, 3, keyword="custom", extra="bonus", num=42)
````

### 3.4 Scope — Local vs Global

````python
# Local scope — variables inside a function are LOCAL
def my_function():
    local_var = "I only exist inside the function"
    print(local_var)

my_function()
# print(local_var)   # ❌ NameError — doesn't exist outside

# Global scope — variables outside are GLOBAL
global_var = "I exist everywhere"

def read_global():
    print(global_var)   # can READ globals

read_global()
````

````python
# Modifying globals — use the global keyword (use sparingly!)
counter = 0

def increment():
    global counter   # tell Python we mean the GLOBAL counter
    counter += 1

print("Before:", counter)
increment()
increment()
increment()
print("After:", counter)   # 3

# Better practice: use return instead of modifying globals
def increment_better(count):
    return count + 1

counter2 = 0
counter2 = increment_better(counter2)
counter2 = increment_better(counter2)
print("Counter2:", counter2)
````
