## 1. What is Inheritance?

Inheritance lets a **child class** acquire the attributes and methods of a **parent class**, so you don't repeat yourself.

```
Animal  ← parent (base class)
  ├── Dog
  ├── Cat
  └── Bird
```

> **Rule of thumb:** Use inheritance when you can say *"Child IS-A Parent"*  
> e.g., "A Dog IS-A Animal" ✅


````python
# Parent class
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

    def eat(self):
        print(f"{self.name} is eating. 🍖")

    def sleep(self):
        print(f"{self.name} is sleeping. 💤")

    def __str__(self):
        return f"{self.__class__.__name__}(name='{self.name}', age={self.age})"


# Child class — inherits everything from Animal
class Dog(Animal):
    def bark(self):
        print(f"{self.name} says: Woof! 🐶")

class Cat(Animal):
    def meow(self):
        print(f"{self.name} says: Meow! 🐱")

d = Dog("Buddy", 3)
c = Cat("Whiskers", 5)

d.eat()    # inherited from Animal
d.bark()   # Dog's own method
c.sleep()  # inherited
c.meow()   # Cat's own method
print(d)
````

## 2. Using `super()` — Extending the Parent

`super()` lets the child call the parent's methods, especially useful in `__init__`.

````python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)   # call Animal's __init__
        self.breed = breed            # add Dog-specific attribute

    def describe(self):
        print(f"{self.name} | {self.breed} | {self.age} yrs old")

class Cat(Animal):
    def __init__(self, name, age, indoor):
        super().__init__(name, age)
        self.indoor = indoor

    def describe(self):
        location = "indoor" if self.indoor else "outdoor"
        print(f"{self.name} | {self.age} yrs old | {location} cat")

d = Dog("Buddy", 3, "Labrador")
c = Cat("Whiskers", 5, True)
d.describe()
c.describe()
````

## 3. Method Overriding

A child class can **override** (replace) any method from the parent.

````python
#Parent Class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):                        # generic version
        print(f"{self.name} makes a sound.")

class Dog(Animal):
    def speak(self):                        # Dog-specific override
        print(f"{self.name} says: Woof! 🐶")

class Cat(Animal):
    def speak(self):
        print(f"{self.name} says: Meow! 🐱")

class Duck(Animal):
    def speak(self):
        print(f"{self.name} says: Quack! 🦆")

class Fish(Animal):
    pass                                    # no override — uses Animal.speak

animals = [Dog("Buddy"), Cat("Whiskers"), Duck("Donald"), Fish("Nemo")]
for a in animals:
    a.speak()
````

## 4. Polymorphism

Polymorphism means *many forms*. The same method call behaves differently depending on the object — as seen above with `speak()`. This is powerful because your code can work with any animal without knowing its exact type.

````python
# Polymorphism in action — a function that works with ANY Animal subclass
class Animal:
    def __init__(self, name): self.name = name
    def speak(self): raise NotImplementedError

class Dog(Animal):
    def speak(self): return f"{self.name}: Woof!"

class Cat(Animal):
    def speak(self): return f"{self.name}: Meow!"

class Bird(Animal):
    def speak(self): return f"{self.name}: Tweet!"

def make_noise(animal):     # works for any animal!
    print(animal.speak())

pets = [Dog("Rex"), Cat("Luna"), Bird("Tweety"), Dog("Max")]
for pet in pets:
    make_noise(pet)
````

## 5. `isinstance()` and `issubclass()`

````python
class Animal: pass
class Dog(Animal): pass
class Cat(Animal): pass

d = Dog()

print(isinstance(d, Dog))     # True  — d is a Dog
print(isinstance(d, Animal))  # True  — Dog IS-A Animal
print(isinstance(d, Cat))     # False

print(issubclass(Dog, Animal))   # True
print(issubclass(Cat, Animal))   # True
print(issubclass(Dog, Cat))      # False
````

## 6. Abstract Base Classes (ABCs)

An **abstract class** defines a blueprint that child classes *must* follow. You cannot instantiate it directly.

````python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    def describe(self):   # concrete method — shared by all
        print(f"Area: {self.area():.2f} | Perimeter: {self.perimeter():.2f}")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):      return 3.14159 * self.radius ** 2
    def perimeter(self): return 2 * 3.14159 * self.radius

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w, self.h = w, h
    def area(self):      return self.w * self.h
    def perimeter(self): return 2 * (self.w + self.h)

shapes = [Circle(5), Rectangle(4, 6)]
for s in shapes:
    s.describe()

# This would raise TypeError:
# s = Shape()
````

## 7. 🧪 Mini Exercises

````python
# Exercise: Build a Vehicle hierarchy
# Parent: Vehicle(make, model, year) → method: info()
# Child:  Car(make, model, year, doors) → override info(), add honk()
# Child:  Truck(make, model, year, payload_tons) → override info(), add load()
# Child:  ElectricCar(make, model, year, doors, battery_kwh) → override info()

class Vehicle:
    
    def __init__(self, brand:str, model:str, year:int):
        self.brand = brand
        self.model = model
        self.year = year

    def info(self):
        print(f"I'm calling from Vehicle : Brand = {self.brand} | Model = {self.model} | Year = {self.year}")


class Car(Vehicle):

    def __init__(self, brand, model, year, doors):
        super().__init__(brand, model, year)
        self.doors = doors

    def info(self):
        print(f"I'm calling from Car : Brand = {self.brand} | Model = {self.model} | Year = {self.year} | Doors = {self.doors}")

    def honk(self):
        print(f"The car is making a sound!")

class Truck(Vehicle):

    def __init__(self, brand, model, year, payload_tons):
        super().__init__(brand, model, year)
        self.payload_tons = payload_tons

    def info(self):
        print(f"I'm calling from Truck : Brand = {self.brand} | Model = {self.model} | Year = {self.year} | Payload Tons = {self.payload_tons}")

    def load(self, load):
        self.payload_tons += load
        print(f"The current payload is {self.payload_tons}")


class ElectricCar(Vehicle):
    
    def __init__(self, brand, model, year, doors, battery_kwh):
        super().__init__(brand, model, year)
        self.doors = doors
        self.battery_kwh = battery_kwh





# Tests
v  = Vehicle("Generic", "Model", 2020)
c  = Car("Toyota", "Camry", 2022, 4)
t  = Truck("Ford", "F-150", 2021, 1.5)
e  = ElectricCar("Tesla", "Model 3", 2023, 4, 500)

for obj in [v, c, t, e]:
    obj.info()
````

## ✅ Summary

| Concept | Key Point |
|---|---|
| Inheritance | Child class gets all parent attributes & methods |
| `super()` | Call parent's `__init__` or any method |
| Method Override | Child redefines a parent method |
| Polymorphism | Same method name, different behaviour per class |
| `isinstance()` | Check if object is of a class (or subclass) |
| Abstract class | Blueprint — forces child classes to implement methods |
