## 1. Opening Files — The `open()` Function

```python
file = open("filename.txt", mode)
```

| Mode | Meaning | Creates file? | Overwrites? |
|------|---------|--------------|-------------|
| `'r'` | Read (default) | ❌ | ❌ |
| `'w'` | Write | ✅ | ✅ |
| `'a'` | Append | ✅ | ❌ |
| `'x'` | Create (fails if exists) | ✅ | ❌ |
| `'r+'` | Read + Write | ❌ | ❌ |

> ⚠️ Always close files after use. The `with` statement does this automatically.


````python
f = open("hello.txt", "r")   # open for reading
````

````python
file = open("files/simple.txt", 'r')
content = file.read()
print(content)
file.close()
````

````python
# Writing to a file
with open("files/simple.txt", "w") as f: #no need to write file.close()
    f.write("Hello, World!\n")
    f.write("This is line 2.\n")
    f.write("Python file handling is easy.\n")

print("File written successfully ✅")
````

````python
# Reading an entire file at once
with open("files/simple.txt", "r") as f:
    content = f.read()

print(content)
print(f"Total characters: {len(content)}")
````

## 2. Reading Files Line by Line

For large files, reading line by line is memory-efficient.

````python
# Method 1: readlines() — returns a list of lines
with open("files/simple.txt", "r") as f:
    line1 = f.readline()
    line2 = f.readline()
    line3 = f.readline()
    line4 = f.readline()

print(line1, line2, line3, line4)

# for i, line in enumerate(lines, 1):
#     print(f"Line {i}: {line.strip()}")

# print(f"\nTotal lines: {len(lines)}")
````

````python
# Method 2: iterate directly (most memory-efficient)
with open("files/simple.txt", "r") as f:
    for line in f:
        print(f"{line}")
````

## 3. Appending to Files

Use `'a'` mode to add content without overwriting existing data.

````python
# Append new lines
with open("sample.txt", "a") as f:
    f.write("This line was appended.\n")
    f.write("So was this one!\n")

# Verify
with open("sample.txt", "r") as f:
    print(f.read())
````

## 4. Writing Multiple Lines with `writelines()`

````python
students = ["Alice\n", "Bob\n", "Charlie\n", "Diana\n"]

with open("students.txt", "w") as f:
    f.writelines(students)

# Read back
with open("students.txt", "r") as f:
    for line in f:
        print(line.strip())
````

## 5. Working with File Paths — `os` and `pathlib`

````python
# cd - change directory
# cwd - current working directory
# ls - list items
````

````python
import os

# Current working directory
print("CWD:", os.getcwd())

# Check if file exists
print("sample.txt exists:", os.path.exists("files/simple.txt"))
print("missing.txt exists:", os.path.exists("missing.txt"))

# File info
print("Size (bytes):", os.path.getsize("files/simple.txt"))
print("Basename:", os.path.basename("files/simple.txt"))
print("Directory:", os.path.dirname("files/simple.txt"))
````

````python
from pathlib import Path

p = Path("files/simple.txt")
print("Exists:", p.exists())
print("Suffix:", p.suffix)
print("Stem:  ", p.stem)
print("Size:  ", p.stat().st_size, "bytes")

# Read with pathlib (no open() needed!)
content = p.read_text()
print("\nContent via pathlib:")
print(content)
````

## 6. Exception Handling for Files

````python
files = {'file_1' : "files/simple.txt", 'file_2' : "files/sample.txt"}

try:
    files['file_3']
except KeyError as e:
    print(f"There is a key error in the above code - {e}")
except Exception as e:
    print(f"There is some error in the above code - {e}")
````

````python
# Handle missing files gracefully
def safe_read(filepath):
    try:
        with open(filepath, "r") as f:
            return f.read()
    except FileNotFoundError:
        return f"❌ Error: '{filepath}' not found."
    except PermissionError:
        return f"❌ Error: No permission to read '{filepath}'."
    except Exception as e:
        return f"❌ Unexpected error: {e}"

print(safe_read("sample.txt"))
print(safe_read("ghost_file.txt"))
````
