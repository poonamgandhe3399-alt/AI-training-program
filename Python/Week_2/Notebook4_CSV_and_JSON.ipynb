## Part A — CSV Files

A **CSV (Comma-Separated Values)** file stores tabular data as plain text.

```
name,age,grade
Alice,20,A
Bob,22,B
Charlie,21,A+
```

Python's built-in `csv` module handles reading and writing.


````python
import csv

# Writing a CSV file
students = [
    ["name",    "age", "grade", "score"],
    ["Alice",   20,    "A",     92],
    ["Bob",     22,    "B",     78],
    ["Charlie", 21,    "A+",    97],
    ["Diana",   23,    "B+",    85],
    ["Ethan",   20,    "C",     65],
]

try:
    with open("students.csv", "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerows(students)
except Exception as e:
    print(f"{e}")

print("CSV written ✅")
````

````python
class Animal:

    def __init__(self, name):
        self.name = name

animal = Animal('Dog')
print(animal)
````

````python
# Reading a CSV file — csv.reader
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    # print(reader)
    for row in reader:
        print(row)
````

````python
# Better: csv.DictReader — each row becomes a dict
with open("students.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row)
        print(f"  {row['name']:10s} | Age: {row['age']} | Grade: {row['grade']} | Score: {row['score']}")
        print('====')
````

## 2. Writing Dicts to CSV with `DictWriter`

````python
import csv

records = [
    {"name": "Fiona",  "subject": "Maths",   "marks": 88},
    {"name": "George", "subject": "Science",  "marks": 74},
    {"name": "Hannah", "subject": "English",  "marks": 91},
]

with open("marks.csv", "w", newline="") as f:
    fieldnames = ["name", "subject", "marks"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(records)

# Verify
with open("marks.csv", "r") as f:
    print(f.read())
````

## 3. CSV with Pandas 🐼

Pandas makes CSV handling much more powerful.

````python
import pandas as pd

# Load CSV into DataFrame
df = pd.read_csv("students.csv")
print(df)
````

````python
df.describe()
print("\nShape:", df.shape) #(no.of rows, no.of columns)
print("\nColumn types:\n", df.dtypes)
````

````python
# Useful pandas operations
df["score"] = pd.to_numeric(df["score"])
print("Score", df['score'])

print("Average score:", df["score"].mean())
print("Highest score:", df["score"].max())
print("\nTop scorer:")
print(df[df["score"] == df["score"].max()])

````

````python
print("\nStudents scoring above 80:")
print(df[(df["score"] > 80) & (df["score"] < 95)])
````

## Part B — JSON Files

**JSON (JavaScript Object Notation)** is the standard format for APIs and configuration files.

```json
{
  "name": "Alice",
  "age": 20,
  "grades": ["A", "A+", "B"]
}
```

Python's `json` module maps JSON ↔ Python types automatically.

| JSON | Python |
|------|--------|
| object `{}` | `dict` |
| array `[]` | `list` |
| string | `str` |
| number | `int` / `float` |
| true/false | `True` / `False` |
| null | `None` |


````python
import json

# Python dict → JSON string
student = {
    "name":   "Alice",
    "age":    20,
    "grades": {"Maths": 92, "Science": 88, "English": 95},
    "active": True,
    "mentor": None
}

json_string = json.dumps(student, indent=4) # converts student dict to json_string
print(json_string)
print(type(json_string))
````

````python
# JSON string → Python dict
parsed = json.loads(json_string)
print(type(parsed))
print(parsed["name"])
print(parsed["grades"]["Maths"])
````

````python
# Write JSON to a file
with open("student.json", "w") as f:
    json.dump(student, f, indent=4)

print("JSON file written ✅")

# Read JSON from a file
with open("student.json", "r") as f:
    loaded = json.load(f)

print(loaded)
````

## 5. Working with a List of JSON Records

````python
import json

# A mini database as a list of dicts
students_db = [
    {"id": 1, "name": "Alice",   "scores": [92, 88, 95], "year": 2},
    {"id": 2, "name": "Bob",     "scores": [78, 65, 82], "year": 1},
    {"id": 3, "name": "Charlie", "scores": [97, 91, 89], "year": 3},
]

# Save
with open("students_db.json", "w") as f:
    json.dump(students_db, f, indent=4)

# Load and query
with open("students_db.json", "r") as f:
    db = json.load(f)

for s in db:
    avg = sum(s["scores"]) / len(s["scores"])
    print(f"  {s['name']:10s} | Avg: {avg:.1f} | Year: {s['year']}")
````
