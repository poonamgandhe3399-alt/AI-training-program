## 1. What is OOP?

**Object-Oriented Programming** is a way of organising code around *objects* — bundles of **data (attributes)** and **behaviour (methods)**.

| Concept | Real-world analogy | Python |
|---|---|---|
| Class | Blueprint of a house | `class House:` |
| Object | An actual house built from that blueprint | `my_house = House()` |
| Attribute | Number of rooms, colour | `self.rooms = 3` |
| Method | Open door, turn on lights | `def open_door(self):` |

> Think of a **class** as a cookie cutter and **objects** as the cookies.


```python
class Dog:

    def __init__(self, name = 'Tommy', age = 9, color = 'Brown'):
        self.dog_name = name
        self.age = age
        self.color = color

dog1 = Dog()
print(type(dog1))
print(f"Name = {dog1.dog_name}, Age = {dog1.age}, Color = {dog1.color}")
```

```python
# Your first class — the simplest possible
class Dog:
    pass          # 'pass' means the class body is empty for now

# Create two objects (instances) from the class
dog1 = Dog()
dog2 = Dog()

print(type(dog1))   # <class '__main__.Dog'>
print(dog1 is dog2) # False — they are different objects
```

## 2. The `__init__` Method — Giving Objects Attributes

`__init__` is called automatically when you create a new object. Use it to set up the object's initial state.

```python
class Dog:
    # __init__ runs the moment you write Dog(...)
    def __init__(self, name, breed, age):
        self.name  = name    # instance attribute
        self.breed = breed
        self.age   = age

# Create objects by passing arguments
dog1 = Dog("Buddy",   "Labrador", 3)
dog2 = Dog("Charlie", "Poodle",   5)

print(dog1.name,  dog1.breed, dog1.age)
print(dog2.name,  dog2.breed, dog2.age)
```

## 3. Instance Methods — Adding Behaviour

Methods are functions defined inside a class. The first parameter is always `self` — it refers to the object calling the method.

```python
class Dog:
    def __init__(self, name, breed, age):
        self.name  = name
        self.breed = breed
        self.age   = age

    def bark(self):
        print(f"{self.name} says: Woof! 🐶")

    def describe(self, color):
        print(f"I am {self.name}, a {self.age}-year-old {self.breed}.")
        print(f"Color of the dog = {color}")
        print('====')

    def have_birthday(self):
        self.age += 1
        print(f"Happy Birthday {self.name}! You are now {self.age}.")

dog1 = Dog("Buddy", "Labrador", 3)
dog1.bark()
dog1.describe('Brown')
dog1.have_birthday()
dog1.describe('Brown')
```

## 4. `__str__` and `__repr__` — Readable Output

- `__str__` → used by `print()` — human-friendly
- `__repr__` → used in the REPL — developer-friendly

```python
class Dog:
    def __init__(self, name, breed, age):
        self.name  = name
        self.breed = breed
        self.age   = age

    def __str__(self):
        return f"Dog(name='{self.name}', breed='{self.breed}', age={self.age})"

    # def __repr__(self):
    #     return f"Dog('{self.name}', '{self.breed}', {self.age})"

dog = Dog("Buddy", "Labrador", 3)
print(dog)       # calls __str__
dog              # calls __repr__ in a notebook/REPL
```

```python
dog.__str__()
```

## 5. Class Attributes vs Instance Attributes

- **Class attribute**: shared by ALL instances
- **Instance attribute**: unique to each object

```python
class Dog:
    species = "Canis lupus familiaris"  # class attribute — shared

    def __init__(self, name, age):
        self.name = name   # instance attribute — unique
        self.age  = age

d1 = Dog("Buddy", 3)
d2 = Dog("Max",   7)

print(d1.species)        # same for all
print(d2.species)        # same for all
print(Dog.species)       # also accessible via the class

print(d1.name, d2.name)  # different for each
```

## 6. 🧪 Mini Exercises

```python
# Exercise 1: Create a BankAccount class
# Attributes: owner (str), balance (float)
# Methods:
#   deposit(amount)   → adds to balance, prints new balance
#   withdraw(amount)  → subtracts if funds available, else prints "Insufficient funds"
#   __str__           → "Account[owner=..., balance=₹...]"

class BankAccount:
    
    def __init__(self, owner : str, balance : float):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount : float):
        self.balance += amount
        print(f"The updated balance is ${self.balance}")

    def withdraw(self, amount : float):
        if amount <= self.balance:
            self.balance -= amount
            print(f"The updated balance is ${self.balance}")
        else:
            print(f"Insufficient funds! Your account balance is ${self.balance}")

    def __str__(self):
        return f"Account[owner={self.owner}, balance=${self.balance}]"



# Test
acc = BankAccount("Alice", 1000)
acc.deposit(500)
acc.withdraw(200)
acc.withdraw(2000)
print(acc)
```

## ✅ Summary

| Concept | Syntax |
|---|---|
| Define a class | `class MyClass:` |
| Constructor | `def __init__(self, ...):` |
| Instance method | `def method(self):` |
| Create object | `obj = MyClass(...)` |
| Access attribute | `obj.attribute` |
| Call method | `obj.method()` |
| Readable print | `def __str__(self):` |
| Class attribute | Defined at class level, outside `__init__` |
